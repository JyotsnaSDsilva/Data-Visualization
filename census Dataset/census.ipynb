```python
c2 = census[['BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012', 'BIRTHS2013', 'BIRTHS2014','BIRTHS2015']]
c3 = census[['DEATHS2010', 'DEATHS2011', 'DEATHS2012', 'DEATHS2013', 'DEATHS2014', 'DEATHS2015']]

c3.describe()
```

```python
c2.describe()
```

```python
c2.tail()
```

```python
bth_lst = [c2[i].sum() for i in c2.columns]
dth_lst = [c3[i].sum() for i in c3.columns]
```

```python
bth_lst
```

```python
dth_lst
```

```python
fig = plt.figure(figsize = (20, 10))

ax2 = fig.add_subplot()
x = np.arange(0,6)
y1 = bth_lst
y2 = dth_lst

births = ax2.bar(x - 0.3/2, y1,width=0.3, label='Births',color='b')
deaths = ax2.bar(x + 0.3/2, y2,width=0.3, label='Deaths',color='r')

ax2.set_xticklabels(['2009','2010','2011','2012','2013','2014','2015'])
ax2.set_xlabel('Years')
ax2.set_title('Births/Deaths 2011-2015 ')
ax2.legend(loc = "upper right") 

def autolabel(values):
    for v in values:
        height = v.get_height()
        ax2.annotate('{}'.format(height),
                    xy=(v.get_x() + v.get_width() / 2, height),
                    xytext=(0, 10), textcoords="offset points", ha='center')

autolabel(births)
autolabel(deaths)

plt.show()
```

```python
region = census["REGION"].value_counts().sort_index()
region
```

```python
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot()
explode = (0, 0, 0.1, 0)

ax.pie(region.values, explode=explode, labels =region.index,shadow=True,
autopct = '%1.1f%%',textprops = {'fontsize': 15, 'color' : "white"})
ax.set_title("Pie chart for Region count")
ax.legend(loc = "upper left", bbox_to_anchor = (1, 0, 0.5, 1), fontsize = 10, title = "Region Code")
ax.axis('equal')
plt.show()
```

```python
city = census['CTYNAME'].unique()
len(city)
```

```python
state = census["STNAME"].value_counts().sort_index()
state
```

```python
len(state.index)
```

```python
fig = plt.figure(figsize = (19, 9))
ax1 = fig.add_subplot()

x_1 = state.index
y_1 = state.values

for x_, y_ in zip(x_1, y_1):
    ax1.bar(x_, y_, color = "blue" if y_ < y_1.mean() else "red", alpha = 0.5)
    ax1.text(x_, y_ - y_/2, str(y_), horizontalalignment = 'center')

plt.xticks(rotation=70)
ax1.set_xlabel('States')
ax1.set_ylabel('Count')
ax1.set_title("State Count")

plt.show()
```

```python
cty = census["CTYNAME"].value_counts().sort_values(ascending=False)[:30]
cty
```

```python
fig = plt.figure(figsize = (19, 9))
ax1 = fig.add_subplot()

x_1 = cty.index
y_1 = cty.values

for x_, y_ in zip(x_1, y_1):
    ax1.bar(x_, y_, color = "green" if y_ < y_1.mean() else "red", alpha = 0.5)
    ax1.text(x_, y_ - y_/2, str(y_), horizontalalignment = 'center')

plt.xticks(rotation=70)
ax1.set_xlabel('County')
ax1.set_ylabel('Frequency')
ax1.set_title("County - Top 30 ")

plt.show()
```
